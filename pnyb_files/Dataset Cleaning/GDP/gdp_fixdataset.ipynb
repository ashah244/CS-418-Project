## Cleaning GDP Dataset

Import the CSV here (CSV must be stored exactly where this .pnyb file is located)

```python
df = pd.read_csv('../../../datasets/CleanedGDP/Clean_gdp.csv')

# Prints the first 5 from the data
```

```python
def createDFRowValueForYear(myYear):
    yearData = df['Year']==myYear
    df2 = df.loc[yearData].copy()
    df2GDP = df2['GDP per Capita']
    df2GDP = df2GDP.reset_index(drop=True)
    return df2GDP

# Create unique list of Country List
countryList = df['Country'].unique()

# Year List
yearList = [2000, 2005, 2010, 2015]

# Create Empty Dataframe with same columns
newDF = pd.DataFrame(columns=['Country', 2000, 2005, 2010, 2015])

#Append country list
newDF['Country'] = countryList

for year in yearList:
    tempDF = createDFRowValueForYear(year)
    newDF[year] = tempDF

newDF
```

```python
# Save Curated Dataset
newDF.to_csv ('../../../datasets/CleanedGDP/Clean_gdp2.csv', index = False)
```
